```python
result_dict = ad_df.set_index('addresse')[['lon', 'lat']].apply(tuple, axis=1).to_dict()
```

```python
import numpy as np
def samedict(address):
    if type(address) != np.nan:
        try:
            if result_dict[address]:
                return result_dict[address][0],result_dict[address][1]
            else:
                return 'None', 'None'
        except:
            pass

all_df['lon'], all_df['lat'] = zip(*all_df['addresse'].apply(samedict))
```

```python
all_df
```

```python
# 'address'를 기준으로 데이터프레임 합치기
merged_df = pd.merge(ad_df[['addresse', 'lon', 'lat']], all_df[['addresse', 'lon', 'lat']], on='addresse', suffixes=('_1', '_2'))

# 'lon'과 'lat' 열이 모두 NaN이 아닌 경우에 값을 덮어씌우기
merged_df.loc[merged_df['lon_1'].notnull() & merged_df['lon_2'].notnull(), ['lon_1', 'lat_1']] = merged_df.loc[merged_df['lon_1'].notnull() & merged_df['lon_2'].notnull(), ['lon_2', 'lat_2']]

# 결과 출력
print(merged_df[['addresse', 'lon_1', 'lat_1']])
```
